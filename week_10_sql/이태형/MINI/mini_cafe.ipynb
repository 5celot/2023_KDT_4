```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
```

```python
import pandas as pd
# 장소 데이터를 불러 옴
cafe=pd.read_csv(r'data\cafe.csv',header=None)
cafe = cafe.rename(columns={0: '장소', 1: '별점', 2: '주소'})
cafe
```

```python
# 데이터프레임 주소값 추출
address= cafe['주소']
print(address.head())
```

```python
# 주소 데이터 깔끔하게 다듬기
add_list=[]
for i in range(len(address)):
    if len(address[i])>=4:
        a = address[i].split(' ')
        add_list.append(" ".join(a[0:4]))
    else:
        a = address[i].split(' ')
        add_list.append(" ".join(a[0:4]))
add_list
```

```python
cafe['수정된 주소']=add_list
cafe
```

```python
from tqdm import tqdm
latitude = []
longitude =[]

for i in tqdm(add_list):
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
cafe['위도']=latitude
cafe['경도']=longitude

cafe
```

```python
cafe[cafe['위도']==0]

# place.to_csv('jungu_incheon_coffeeshop.csv')
```

```python
# 위도 경도가 0인 주소
address2=cafe[cafe['위도']==0]['수정된 주소']
address2
```

```python
# 주소 데이터 깔끔하게 다듬기
add_list2=[]
for i in address2:
    a =i.split(' ')
    add_list2.append(" ".join(a[:-1]))

add_list2
```

```python
from tqdm import tqdm
latitude2 = []
longitude2 =[]

for i in tqdm(add_list2):
    latitude2.append(geocoding(i)[0])
    longitude2.append(geocoding(i)[1])
```

```python
address2.index
```

```python
for i in range(len(address2.index)):
    index_num = address2.index[i]
    cafe.at[index_num, '수정된 주소'] = add_list2[i]
    cafe.at[index_num, '위도'] = latitude2[i]
    cafe.at[index_num, '경도'] = longitude2[i]
cafe
```

```python
cafe[cafe['위도']==0]
```

```python
cafe=cafe.drop(index=cafe[cafe['위도']==0].index)
```

```python
cafe.reset_index(drop=True,inplace=True)
cafe
```

```python
cafe.to_csv('cafe_location.csv',index=None)
```
