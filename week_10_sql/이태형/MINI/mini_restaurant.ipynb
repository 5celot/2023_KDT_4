```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
```

```python
import pandas as pd
# 장소 데이터를 불러 옴
restaurant=pd.read_csv(r'data\restaurant.csv')
restaurant
```

```python
# 데이터프레임 주소값 추출
address= restaurant['주소']
print(address.head())
```

```python
# 주소 데이터 깔끔하게 다듬기
add_list=[]
for i in range(len(address)):
    if len(address[i])>=4:
        a = address[i].split(' ')
        add_list.append(" ".join(a[0:4]))
    else:
        a = address[i].split(' ')
        add_list.append(" ".join(a[0:4]))
add_list
```

```python
restaurant['수정된 주소']=add_list
restaurant
```

```python
from tqdm import tqdm
latitude = []
longitude =[]

for i in tqdm(add_list):
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
restaurant['위도']=latitude
restaurant['경도']=longitude

restaurant
```

```python
restaurant[restaurant['위도']==0]

# place.to_csv('jungu_incheon_coffeeshop.csv')
```

```python
# 위도 경도가 0인 주소
address2=restaurant[restaurant['위도']==0]['수정된 주소']
address2
```

```python
# 주소 데이터 깔끔하게 다듬기
add_list2=[]
for i in address2:
    a =i.split(' ')
    add_list2.append(" ".join(a[:-1]))

add_list2
```

```python
from tqdm import tqdm
latitude2 = []
longitude2 =[]

for i in tqdm(add_list2):
    latitude2.append(geocoding(i)[0])
    longitude2.append(geocoding(i)[1])
```

```python
address2.index
```

```python
for i in range(len(address2.index)):
    index_num = address2.index[i]
    restaurant.at[index_num, '수정된 주소'] = add_list2[i]
    restaurant.at[index_num, '위도'] = latitude2[i]
    restaurant.at[index_num, '경도'] = longitude2[i]
restaurant
```

```python
restaurant[restaurant['위도']==0]
```

```python
restaurant=restaurant.drop(index=restaurant[restaurant['위도']==0].index)
```

```python
restaurant.reset_index(drop=True,inplace=True)
restaurant
```

```python
restaurant.to_csv('restaurant_location.csv',index=None)
```
