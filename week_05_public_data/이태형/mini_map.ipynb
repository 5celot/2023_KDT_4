```python
import pandas as pd
import folium

# CSV 파일 읽어오기
df = pd.read_csv(r'data\아파트_매매_실거래_평균가격_20230810154226.csv', encoding='utf-8')# 지역별 아파트 매매
df
```

```python
df.drop(df.index[0:3],inplace=True)
df.drop(df.index[1:6],inplace=True)
df.drop('행정구역별(1)',axis=1)
```

```python
df01=df[['행정구역별(2)','2017.05','2023.05']]
df01['10년 평균매매가격 변화률']=round((df01['2023.05']-df01['2017.05'])/(df01['2017.05'])*100,1)
df01=df01[['행정구역별(2)','10년 평균매매가격 변화률']]
df01
```

```python
addr_aliases =	{'서울':'서울특별시',	'부산':'부산광역시',	'대구':'대구광역시',	'인천':'인천광역시',	
'광주':'광주광역시',	'대전':'대전광역시',	'울산':'울산광역시',	'경기':'경기도','소계':'세종특별자치시',
'강원':'강원도',	'충북':'충청북도','제주':'제주특별자치도','충남':'충청남도','전북':'전라북도','전남':'전라남도','경북':'경상북도','경남':'경상남도'}
df01['행정구역별(2)']=	df01['행정구역별(2)'].apply(lambda	v:	addr_aliases.get(v,	v))
df01
```

```python

paldo_map=folium.Map(location=[36,128],zoom_start=8)
geo_path='SIDO_MAP_2022.json' # 시도 경계 데이터

counts = df01['10년 평균매매가격 변화률'].tolist

# 지도에 Choropleth Layer 추가하기
folium.Choropleth(
    geo_data=geo_path,
    data=df01,
    columns=['행정구역별(2)', '10년 평균매매가격 변화률'],
    fill_color='YlOrRd',
    key_on='feature.properties.CTP_KOR_NM',
    vmin=0,  # 값의 최솟값
    vmax=100,  # 값의 최댓값
    fill_opacity=0.8,
    line_opacity=0.5,
).add_to(paldo_map)


# LayerControl 추가하기
folium.LayerControl().add_to(paldo_map)

# 지도를 HTML 파일로 저장하기
paldo_map.save('apart.html')
```
