```python
import pandas as pd
import folium
import json
# CSV 파일 읽어오기
df = pd.read_csv('data_junior.csv', encoding='utf-8')#신입 csv파일
df
```

```python

paldo_map=folium.Map(location=[36,128],zoom_start=8)
geo_path='SIDO_MAP_2022.json' # 시도 경계 데이터

counts = df['지역'].value_counts()

# 지도에 Choropleth Layer 추가하기
folium.Choropleth(
    geo_data=geo_path,
    data=counts,
    columns=[counts.index, counts],
    fill_color='YlOrRd',
    key_on='feature.properties.CTP_KOR_NM',
    vmin=0,  # 값의 최솟값
    vmax=25,  # 값의 최댓값
    fill_opacity=0.8,
    line_opacity=0.5,
).add_to(paldo_map)


# LayerControl 추가하기
folium.LayerControl().add_to(paldo_map)

# 지도를 HTML 파일로 저장하기
paldo_map.save('paldo_map_jonior.html')
```
